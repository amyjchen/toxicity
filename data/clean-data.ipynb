```python
import csv
import math
import pandas as pd
import numpy as np
```

```python
filename = 'train.csv'
```

```python
df = pd.read_csv(filename)
```

```python
df = df.drop([
    'id', 
    'article_id',
    'rating', 
    'funny',
    'wow',
    'sad',
    'likes',
    'disagree', 
    'created_date', 
    'publication_id',
    'parent_id'
], axis = 1)
```

```python
df
```

We see that we have an issue with our data where some strings have no annotators for identity. 
This can pose an issue for us. We split the data into two, one with identities annotated and one without. 

```python
df_annotated = df[df.identity_annotator_count > 0]
df_annotated
```

```python
df_not_annotated = df[df.identity_annotator_count == 0]
df_not_annotated
```

About a fourth of our data had identity annotations, which suggests that we may need to predict annotations on our own. 

For our baseline, we wont include identifying variables (we'll focus on our X (the comment_text) and multiple Y (target, severity of toxicity, obsenity, identity_attacks, insults, and threats). For now we save the other data into files. 

```python
annotated_filename = 'train_id_annotated.csv'
df_annotated.to_csv(annotated_filename, encoding='utf-8', index=False)
not_annotated_filename = 'train_id_not_annotated.csv'
df_not_annotated.to_csv(not_annotated_filename, encoding='utf-8', index=False)
```
